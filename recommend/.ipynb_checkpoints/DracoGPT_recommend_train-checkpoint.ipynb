```python
from sklearn import svm
from sklearn.model_selection import train_test_split
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.model_selection import cross_val_predict
from sklearn.svm import LinearSVC
from sklearn.linear_model import LogisticRegression
from sklearn.calibration import CalibratedClassifierCV

np.random.seed(42)

def train_model(data: pd.DataFrame, test_size: float = 0.2, C_values: list = [0.1, 0.5, 1, 5, 10], quiet=False):
    # Prepare the data
    print("*"*90)
    X = data.negative - data.positive
    size = len(X)
    y = np.ones(size)

    # Generate synthetic labels for demonstration
    idx = np.random.choice([False, True], size=size, p=[0.5, 0.5])
    X[idx] = -X[idx]
    y[idx] = -y[idx]

    # Split the data into train and test sets
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size)

    best_score = 0
    best_C = None

    # Perform 5-fold cross-validation to find the best C
    for C in C_values:
        clf = svm.LinearSVC(C=C, fit_intercept=False, max_iter=10000, dual=False)
        scores = cross_val_score(clf, X_train, y_train, cv=5)  # 5-fold CV
        avg_score = np.mean(scores)
        
        if not quiet:
            print(f"C: {C}, Avg CV Score: {avg_score}")

        if avg_score > best_score:
            best_score = avg_score
            best_C = C

    
    clf_final = svm.LinearSVC(C=best_C, fit_intercept=False, max_iter=10000, dual=False)
    clf_final.fit(X_train, y_train)

    final_test_score = clf_final.score(X_test, y_test)
    print(f"Best C: {best_C}, Final Test Score: {final_test_score}")

    # Retrain on the entire training dataset with the best C
    clf_final = svm.LinearSVC(C=best_C, fit_intercept=False, max_iter=10000, dual=False)
    clf_final.fit(X, y)

    return clf_final
```

```python
# write difference vector to agree/disagree/conflict and prepare training data for weights
def train_data_prep(model, temp, skip_shape = False, skip_two_non_pos = False):
    kim = {}

    with open(f"./training_data/kim2018/{model}_{temp}_training.json") as file:
#     with open(f"./training_data/kim2018/kim2018_draco2.json") as file:

        i = 0
        json_data = json.load(file)
        print(len(json_data))
        for pair in json_data:
            
            if skip_shape and "shape" in str(pair):
                continue
            
            if skip_two_non_pos and "color" in str(pair) and "size" in str(pair):
                continue
            
            
            pair["source"] = "kim"
            pair["pair_id"] = f'{pair["source"]}_{i}'
            kim[pair["pair_id"]] = pair
            i += 1

    combined = kim 
    
#     print(combined["kim_0"])
    
    print("&"*20)
    
    baseline_train_data = pairs_to_vec(combined)
    
    
    
    
    assert set(baseline_train_data.negative.columns) == set(
        default_draco.soft_constraint_names
    ), "Feature names do not match."
    clf = train_model(baseline_train_data)

    baseline_weights = {}
    features = baseline_train_data.negative.columns
    with open(f'./weights/DracoGPT_recommend_{model}_{temp}_weights.txt', 'w') as fout:
        for feature, weight in zip(features, clf.coef_[0]):
            baseline_weights[f"{feature}_weight"] = int(weight * 1000)
            fout.write(f"#const {feature}_weight = {int(weight * 1000)}.\n")
            


```
